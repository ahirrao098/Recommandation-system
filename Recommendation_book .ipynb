```python
book_d=book.iloc[:,0:]
```

```python
#df.rename(columns={"OldName":"NewName"})

book_d=book_d.rename(columns={'User.ID':'userID'})
```

```python
book_d=book_d.rename(columns={'Book.Rating':'rating'})
```

```python
book_d.drop_duplicates(subset=None,keep='first',inplace=False)
```

```python
np.where(book_d.index.duplicated())
```

```python
#user_book_df = book_DF.pivot_table(index='userid',
                                 #columns='book',
                                 #values='rating').reset_index(drop=True)

user_data=book_d.pivot_table(index='userID',columns='Book',values='rating').reset_index(drop=True)
```

```python
user_data.head()
```

```python
user_data.index=book_d.userID.unique()#rearrange the data
```

```python
user_data
```

```python
#Impute those NaNs with 0 values
user_data.fillna(0, inplace=True)
```

```python
user_data
```

```python
#Calculating Cosine Similarity between Users
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
user_sim=1-pairwise_distances(user_data.values,metric='cosine')#cosine base simillarity
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index =book_d.userID.unique()
user_sim_df.columns =book_d.userID.unique()
```
